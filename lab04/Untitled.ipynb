```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 500
```

```python
temp = np.loadtxt("/home/nick/Documents/lab 04/scope_7.csv",skiprows=2,delimiter=',')
print(temp)
print(temp.shape)
in_signal = np.vstack((temp[:,0],temp[:,1]))
out_signal = np.vstack((temp[:,0],temp[:,2]))
print(temp[:,1])
```

# A

```python
print(in_signal)
```

```python
plt.scatter(in_signal[0],in_signal[1],label="Input signal")
plt.scatter(out_signal[0],out_signal[1],label="Output Signal")
plt.xlim(-0.0011,0.0011)
plt.xticks(np.arange(min(in_signal[0]),max(in_signal[0]),0.0003))
plt.xlabel("Seconds")
plt.ylabel("Volts")
plt.title("Inverting Amplifier")
legend = plt.legend()
plt.show()
```

```python
gain_t = out_signal[1,:]//in_signal[1,:]
print(gain_t)
plt.yscale('symlog')
plt.ylim((min(gain_t),max(gain_t)+1))
print(max(gain_t))
plt.scatter(in_signal[0,:],gain_t)
plt.xlim(min(in_signal[0,:]),max(in_signal[0,:]))
plt.ylim((min(gain_t),max(gain_t)+1))
plt.xticks(np.arange(min(in_signal[0]),max(in_signal[0]),0.0003))
plt.xlabel("Seconds")
plt.ylabel("Gain")
plt.title("Inverting Amplifier Gain")
```
