```python


import numpy as np
import math
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
from decimal import *
mpl.rcParams['figure.dpi'] = 500

```

```python
def plot_osc_2_scale(f_name,title,col_1="",col_2="",flipped=False,percent_shown=100):
    temp = np.loadtxt(f_name,skiprows=4,delimiter=',')
    #print(temp)
    print(temp.shape)
    size = temp.shape[0]
    size_plot=math.floor(size*(percent_shown/100.0))
    in_signal = np.vstack((temp[:,0],temp[:,1]))
    out_signal = np.vstack((temp[:,0],temp[:,2]))
    if(flipped==True):
        in_signal= np.vstack((temp[:,0],temp[:,2]))
        out_signal= np.vstack((temp[:,0],temp[:,1]))
    #print(temp[:,1])
    
    fig, ax1 = plt.subplots()
    ax1.set_xlabel("time (s)")
    ax1.set_ylabel("Input signal (V)",color="tab:blue")
    ax1.plot(in_signal[0,0:size_plot],in_signal[1,0:size_plot],label=col_1,color="tab:blue")
    
    
    ax2=ax1.twinx()
    ax2.plot(out_signal[0,0:size_plot],out_signal[1,0:size_plot],label=col_2,color="tab:red")
    #plt.xlim(min(in_signal[0,:]),max(in_signal[0,:]))
    ticks_dist = -(min(in_signal[0])-max(in_signal[0]))/5.0
   # plt.xticks(np.arange(min(in_signal[0]),max(in_signal[0]),ticks_dist))
    ax2.set_ylabel("Output signal (V)",color="tab:red")
    #plt.xlabel("Seconds")
    #plt.ylabel("Volts")
    plt.title(title)
    #legend = plt.legend()
    
    fig.tight_layout()
    plt.show()
def plot_osc_many(f_name,title,col_1="",col_2="",flipped=False,percent_shown=100,labels=""):
    temp = np.loadtxt(f_name,skiprows=4,delimiter=',')
    #print(temp)
    print(temp.shape)
    size = temp.shape[0]
    size_plot=math.floor(size*(percent_shown/100.0))
    in_signal = np.vstack((temp[:,0],temp[:,1]))
    out_signal = np.vstack((temp[:,0],temp[:,2]))
    if(flipped==True):
        in_signal= np.vstack((temp[:,0],temp[:,2]))
        out_signal= np.vstack((temp[:,0],temp[:,1]))
    #print(temp[:,1])
    
    fig, ax1 = plt.subplots()
    ax1.set_xlabel("time (s)")
    ax1.set_ylabel("Input signal (V)",color="tab:blue")
    ax1.plot(in_signal[0,0:size_plot],in_signal[1,0:size_plot],label=col_1,color="tab:blue")
    
    
    ax2=ax1.twinx()
    for i in range(2,temp.shape[1]):
        temp_label=""
        if(labels!=""):
            temp_label=labels[i-2]
        out_signal = np.vstack((temp[:,0],temp[:,i]))
        ax2.plot(out_signal[0,0:size_plot],out_signal[1,0:size_plot],label=labels[i-2])
    #plt.xlim(min(in_signal[0,:]),max(in_signal[0,:]))
    ticks_dist = -(min(in_signal[0])-max(in_signal[0]))/5.0
   # plt.xticks(np.arange(min(in_signal[0]),max(in_signal[0]),ticks_dist))
    ax2.set_ylabel("Output signal (V)",color="tab:red")
    #plt.xlabel("Seconds")
    #plt.ylabel("Volts")
    plt.title(title)
    legend = ax2.legend()
    
    fig.tight_layout()
    plt.show()
```
